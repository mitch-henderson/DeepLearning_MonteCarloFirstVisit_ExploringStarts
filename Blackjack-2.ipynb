# OpenAI's gym Blackjack Environment

This is taken from the description of the environment

>    Blackjack is a card game where the goal is to obtain cards that sum to as
    near as possible to 21 without going over.  They're playing against a fixed
    dealer.
    Face cards (Jack, Queen, King) have point value 10.
    Aces can either count as 11 or 1, and it's called 'usable' at 11.
    This game is placed with an infinite deck (or with replacement).
    The game starts with dealer having one face up and one face down card, while
    player having two face up cards. (Virtually for all Blackjack games today).

>    The player can request additional cards (hit=1) until they decide to stop
    (stick=0) or exceed 21 (bust).

>    After the player sticks, the dealer reveals their facedown card, and draws
    until their sum is 17 or greater.  If the dealer goes bust the player wins.

>    If neither player nor dealer busts, the outcome (win, lose, draw) is
    decided by whose sum is closer to 21.  The reward for winning is +1,
    drawing is 0, and losing is -1.

>    The observation of a 3-tuple of: the players current sum,
    the dealer's one showing card (1-10 where 1 is ace),
    and whether or not the player holds a usable ace (0 or 1).

>    This environment corresponds to the version of the blackjack problem
    described in Example 5.1 in Reinforcement Learning: An Introduction
    by Sutton and Barto.
>   http://incompleteideas.net/book/the-book-2nd.html

## Usable Ace

An Ace in blackjack can have different values, either a `1` or a `11`.  If I had drawn an Ace, and I had a `5` in my hand, then the possible values for any hand would be either `6` or `16`.

If I happen to draw another Ace, can I still have a usable Ace?

Yes, because the first Ace can take on the value of `1` or `11`. That is, after the second Ace my score could still be `7` or `17`.

If my hand gets another `5`, then I would not have a usable Ace anymore, since my score should not go above `21`, so in this case my only possible score would be `5+5+1+1` or `12`. None of my Aces can take on the value of `11` otherwise my hand would go above `21`.

### Playing

````python
HIT = 1
STICK = 0
ACTIONS = [STICK, HIT]
````

````python
def play(env, policy, initial_action=None):
    """play: plays through one game of blackjack
    env: game environment
    policy: player's policy
    initial action: if we want a particular player action
    
    Note: if the initial_action is not None, then the initial 
    action will overide the policy, i.e. the action will 
    be based on the initial_action, the subsequent actions 
    will be based on the policy
    """
    trajectory = []
    
    #Players turn
    while True:
        if initial_action:
            action = initial_action
            initial_action = None
        else:
            action = policy(env._get_obs())
        
        #save the state (prior to taking the action), and the action
        trajectory.append((env._get_obs(),action))
        
        # if action == STICK: break 
        state, reward, done, _ = env.step(action)
        if done:
            return state, reward, trajectory
        

            
````

````python
def dealer_policy(hand):
    if hand < 17:
        return HIT
    return STICK

def player_policy(observable):
    hand, dealer, usable_ace = observable
    if hand < 17:
        return HIT
    return STICK
````

````python
env.seed(42)
print('Initial state:', env.reset())
print('Playing one game...')
play(env, player_policy)
````

# Monte carlo on policy

We use the `play` function above to run through an episode of Monte Carlo Decision Process (MDP). Each game of blackjack is an episode. Because games finish, episodes are finite.

The `gym` environment already sets a reward for the final states: `+1` for winning, `-1` for losing, and `0` for a draw. All the rewards while playing the game are zero. This means that the reward going to non-terminal states is zero. We don't use a discount factor ($\gamma = 1$), therefore the terminal rewards is also the returns for the state.

When the player has an ace in hand, he/she can count it as a `11` or as a `1`. If he/she can hold the ace as a `11` without going bust we say that the ace is **usable**. When it is usable, we count it as a `11` because counting as a `1` means there is no decision to be made because the player should always hit.

The player's decison is based on three variables:

- his/her current sum(12-21)
- the dealer's showing card (1-10)
- whether she holds a usabe ace

Therefore the total number of observable states is `10x10x2 = 200` states.

Consider a player policy in which the player:

* sticks when his/her hand is `17` or more,
* otherwise the player always hits. 

To find the state-value function on this policy via Monte Carlo, we simulate many episodes using this policy. We keep track of the state which is defined by the player's current sum, the dealer's showing card, and whether there is a usable ace for the player. The episode will contain a possible trajectory of state, action, reward. Note that the next `gym` environment returns after an action is taken.


![First-visit MC prediction, for estimating V ≈ vπ ](FirstVisitMCPrediction.jpg)

The trajectory of an episode is shown as $S_0, A_0, R_1, S_1, A_1, R_2, \dots, R_T$. For each trajectory we sum over the $R_{t+1}, R_{t+2}, \dots, R_T$ to calculate the return $G$ for a state $S_t$, assuming no discount factor, $\gamma = 1$.

Since the intermediate rewards are `0`, i.e. $R_{t+1} = 0$, the summation for a particular state $S_t$ is just the reward upon reaching the terminal statem $R_T$ for all the states in the trajectory.

Thus in the code below, we assign the terminal reward returned by `play`, and assign the reward to each of the states in the trajectory. Note that becasue the state space is from $([12,21],[1,10],[0,1])$, we subtract `12` from the first number in the tuple, and a `1` from the second number in the tuple, so that we can use these as the indices into the zero-index numpy arrays.

Note that the algorithm in the above box appends $G$ to the returns array, for a particular state $S_t$, and then averages $G$ in the array for that state. Instead we will keep running a sum of returns for a state $S_t$, and a separate count of how many times we've received a return and divide the sum by the count to arrive at an average.

````python
def monte_carlo_on_policy(num_episodes):
    states_usable_ace = np.zeros((10,10))
    states_usable_ace_count = np.ones((10,10))
    states_no_usable_ace = np.zeros((10,10))
    states_usable_ace_count = np.ones((10,10))
    for i in tqdm(range(num_episodes)):
        env.reset()
        _, reward, trajectory = play(env, player_policy)
        for obs, action in trajectory:
            player_sum, dealer_card, usable_ace = obs
            player_sum -= 12
            dealer_card -= 1
            if usable_ace:
                states_usable_ace[player_sum, dealer_card] += reward
                states_usable_ace_count[player_sum, dealer_card] += 1
            else: 
                states_no_usable_ace[player_sum, dealer_card] += reward
                states_no_usable_ace[player_sum, dealer_card] += 1        
    return states_usable_ace / states_usable_ace_count, states_no_usable_ace / states_no_usable_ace[0]
````

````python
def run_monte_carlo_on_policy():
    episodes = [10000, 500000]
    usable1, non_usable1 = monte_carlo_on_policy(episodes[0])
    usable2, non_usable2 = monte_carlo_on_policy(episodes[1])
    states = [usable1, usable2, non_usable1, non_usable2]
    titles = [
        f'Usable Ace, {episodes[0]} episodes',
        f'Usable Ace, {episodes[1]} episodes',
        f'Non-usable Ace, {episodes[0]} episodes',
        f'Non-usable Ace, {episodes[1]} episodes'
    ]
    return episodes, states, titles
````

# Monte Carlo Control

## Monte Carlo Exploring Starts (ES)

![Monte Carlo with Exploring Starts](monte_carlo_es.jpg)

 # q(s,a)
    
Here we encode the state in the first dimensions and the action in the final dimension. Note the state is 10x10x2, corresponding to:
- player's sum: 12-21
- dealer's showing card: 1-10
- whether the player has a usable ace: Yes(1)/No(0)
- And for the action, since there are only two actions, (hit/stick)
    
Note the return for a trajectory happens to be the terminal reward R_T because the intermediate, non-terminal, rewards are zero, amd discount factor, gamma, is 1. Thus we do not need to loop through the trajectory backwards, i.e. for  ... t = T-1, T-2, ... 0
        
And we do not need to accumulate the retun G incrementally, i.e. G <- gamma G + R {t+1}

````python
def monte_carlo_es(num_episodes):

    state_action_pairs = np.zeros((10, 10, 2, 2))
    state_action_pairs_count = np.zeros((10, 10, 2, 2))
    
    def pi_(state):
        player_sum, dealer_card, usable_ace = state
        # Subtracting to shift the state into array indices
        player_sum -= 12
        dealer_card -= 1
        usable_ace = int(usable_ace)
        
        # print(f"{player_sum="} {dealer_card=} {usable_ace}")
        # print(state_action_pairs[player_sum, dealer_card, usable_ace,:])
        
        q_sa = (
            state_action_pairs[player_sum, dealer_card, usable_ace, :] /
            state_action_pairs_count[player_sum, dealer_card, usable_ace, :]
        )
        a_star = np.flatnonzero(q_sa == np.max(q_sa))
        if len(a_star) == 0:
            return np.random.choice(ACTIONS)
        return np.random.choice(a_star)
        
    for i in tqdm(range(num_episodes)):
        env.reset()
        action = np.random.choice(ACTIONS)
        _, reward, trajectory = play(env, pi_, action)
        first_visit = set()
        for obs, action in trajectory:
            player_sum, dealer_card, usable_ace = obs
            player_sum -= 12
            dealer_card -= 1
            usable_ace = int(usable_ace)
            state_action = (player_sum, dealer_card, usable_ace, action)
            if state_action in first_visit:
                continue
            first_visit.add(state_action)
            state_action_pairs[state_action] += reward
            state_action_pairs_count[state_action] += 1
            
    return state_action_pairs / state_action_pairs_count
            
````

````python
def run_monte_with_exploring_starts(num_episodes):
    # num_episodes = 5000000
    q_ = monte_carlo_es(num_episodes)
    
    # take the max state_action over the actions (index = 3 (or -1 for last))
    v_no_usable_ace = np.max(q_[:,:,0,:], axis=-1)
    v_usable_ace = np.max(q_[:,:,1,:], axis=-1)
    
    # get the optimal policy pi
    action_no_usable_ace = np.argmax(q_[:,:,0,:], axis=-1)
    action_usable_ace = np.argmax(q_[:,:,1,:], axis=-1)
    
    policy_values = [
        action_usable_ace,
        v_usable_ace,
        action_no_usable_ace,
        v_no_usable_ace
    ]
    titles = [
        'Optimal policy with usable Ace',
        'Optimal value with usable Ace',
        'Optimal policy without usable Ace',
        'Optimal value without usable Ace'
    ]
    
    return num_episodes, policy_values, titles
````
